```python
import requests
from bs4 import BeautifulSoup
import re
import numpy as np
import pandas as pd
```

```python
%%time
# The code should complete in a minute. 

# Save each recipe information as a list in the list recipes. 
data = []
URL = "https://en.wikipedia.org/wiki/List_of_multiple_Olympic_medalists#In_the_Summer_and_Winter_Games"
page = requests.get(URL)
soup = BeautifulSoup(page.content, "html.parser")

tables = soup.find_all('table')

dfs = []
for table in tables:
    dfs.append(pd.read_html(str(table))[0])
```

```python
data = dfs[0].copy()
```

```python
data = data.dropna(axis = 0)
data
```

```python
data.to_csv('project_data.csv', index=False)
```

```python
data = pd.read_csv('project_data.csv')
```

```python
data['Years'].unique()
```

```python
data
```

```python
data.drop(labels='No.', axis=1, inplace=True)
data
```

```python
# Extract the first year from the 'Years' column
data['First Medal Year'] = data['Years'].str.extract('^(\d{4})')

# Extract the last year from the 'Years' column
data['Last Medal Year'] = data['Years'].str.extract('(\d{4})[^\d]*$')
```

```python
data['Last Medal Year'].unique()
```

```python
data['Last Medal Year'].isnull().sum()
```

```python
data['First Medal Year'] = data['First Medal Year'].astype(int)
data['Last Medal Year'] = data['Last Medal Year'].astype(int)
```

```python
data.info()
```

```python
data.head(5)
```

```python
data = data.iloc[:,[0,1,2,10,11,4,5,6,7,8,9]].copy()
```

```python
data
```

```python
data['Nation'].unique()
```

```python
country_corrections = {
    "East Germany \xa0Germany": "East Germany",
    "Unified Team \xa0Belarus": "Unified Soviet Team",
    "Soviet Union \xa0Unified Team": "Soviet Union",
    "Unified Team \xa0Russia": "Unified Soviet Team",
    "Olympic Athletes from Russia \xa0ROC": "Russia/ROC",
    "United Team of Germany \xa0West Germany": "West Germany",
    "Russia \xa0ROC": "Russia/ROC",
    "Soviet Union \xa0Kazakhstan": "Soviet Union",
    "Romania \xa0Romania": "Romania",
    "Germany \xa0Germany": "Germany",
    "Soviet Union \xa0Unified Team \xa0Russia": "Soviet Union",
    "Unified Team \xa0Ukraine": "Unified Soviet Team",
    "Australasia \xa0Australia": "Australia",
}

data['Nation'] = data['Nation'].replace(country_corrections)
```

```python
data
```

```python
sport_correction = {
    "Cross-Country skiing Nordic combined": "Nordic combined"
}

data['Sport'] = data['Sport'].replace(sport_correction)
```

```python
data.to_csv('project_data_final.csv', index=False)
```
